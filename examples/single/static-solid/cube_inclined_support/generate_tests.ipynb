```python
from pathlib import Path
import shutil
import os
import subprocess
import numpy as np
```

```python
def find_and_replace(file_path, find_text, replace_text):
    # Open the file for reading
    with open(file_path, 'r') as file:
        # Read the contents of the file
        content = file.read()
    
    # Replace the target text with the desired text
    new_content = content.replace(find_text, replace_text)
    
    # Open the file for writing and save the new content
    with open(file_path, 'w') as file:
        file.write(new_content)
```

```python
angles = [ 0, 22.5, 45, 45+22.5, 90 ]
for i, angle in enumerate(angles):
    journal = f'cube-test{i+1}.jou'
    yaml = f'cube-test{i+1}.yaml'
    mesh = f'cube-test{i+1}.g'
    shutil.copy2('cube-template.jou', journal)
    shutil.copy2('cube-template.yaml', yaml)
    replacements = {
        '{input_mesh_file}': mesh,
        '{cosangle}': str(np.cos(angle*np.pi/180)),
        '{sinangle}': str(np.sin(angle*np.pi/180)),
        '{deg_angle}': str(angle)
    }
    for r in replacements:
        v = replacements[r]
        find_and_replace(journal, r, v )
        find_and_replace(yaml, r, v )
        os.system(f"/Applications/Cubit-16.18/Cubit.app/Contents/MacOS/cubit -batch {journal}")
```
